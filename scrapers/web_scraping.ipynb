```python
from movie_scrape import *
%matplotlib inline
```

```python
lotr_trilogy = ["fellowshipofthering","twotowers","returnoftheking"]
lotr_data = make_series_data(lotr_trilogy)

base_title = "backtothefuture"
btf_trilogy =[base_title]+ [base_title+ str(i) for i in range(1,4) if i>1]
btf_data = make_series_data(btf_trilogy)

s1_label = 'The Lord of the Rings'
s2_label = 'Back to the Future'


fig, ax = plt.subplots()
lotr_plot = makeplot_grosses(lotr_data)
btf_plot = makeplot_grosses(btf_data)
plt.title('Domestic Grosses for Popular Trilogies', fontsize=17);
# plt.xlabel("Films", fontsize=15, labelpad=5);
plt.axis([0.8,3.2,0,600])
plt.xticks(rotation=60)
plt.tick_params(axis='x', which='major', labelsize=15)
plt.ylabel("Inflation adjusted Millions of $", fontsize=15, labelpad=15);
plt.setp(ax, xticks=[1, 2, 3], xticklabels=['1st Film', '2nd Film', '3rd Film']);
plt.setp(btf_plot, color='c', alpha=0.7, linewidth=3.0)
plt.setp(lotr_plot, color='m',alpha=0.7, linewidth=3.0)
plt.legend([s1_label, s2_label],loc='lower left')




plt.show()
```
