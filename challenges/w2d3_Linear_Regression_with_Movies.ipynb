```python
movies_without_budget = movies[ np.isnan(movies.Budget) ]
movies_without_budget
```

```python
movies_with_budget = movies.dropna()
movies_with_budget.tail()
```

```python
movies_with_budget['Ones'] = 1.0
movies_with_budget.tail()
```

```python
import statsmodels.formula.api as sm

X = movies_with_budget[['Budget', 'Ones']]
Y = movies_with_budget['DomesticTotalGross']
linmodel = sm.OLS(Y, X).fit()
linmodel.summary()
```

```python
plt.scatter(movies_with_budget.Budget, movies_with_budget.DomesticTotalGross)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget.Budget, linmodel.predict(), 'r')
```

```python
# Same model with scikitlearn

from sklearn.linear_model import LinearRegression

X = movies_with_budget[['Budget']]
Y = movies_with_budget['DomesticTotalGross']

sk_linmodel = LinearRegression()
sk_linmodel.fit(X,Y)

print 'coefficient:\t %g'% sk_linmodel.coef_
print 'intercept:\t %g'  % sk_linmodel.intercept_
print 'R^2:      \t %g'  % sk_linmodel.score(X, Y)

plt.scatter(X, Y)
plt.xlabel("Budget")
plt.ylabel("Box Office")
plt.plot(X, sk_linmodel.predict(X))
```

#### Plot the predictions of the model against true values

```python
plt.plot(Y, sk_linmodel.predict(X), 'go')
plt.title('Over/Underprediction of the Budget Model')
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Model')

# 1:1 line
plt.plot(Y, Y, 'm--', label = 'Ideal Prediction')
plt.legend()
```

### Multiple Linear Regression
Add a couple of explanatory variables

```python
movies_with_budget['TitleLength'] = [len(title) for title in movies_with_budget.Title]
movies_with_budget.head()
```

```python
X = movies_with_budget[['Budget', 'Runtime', 'TitleLength', 'Ones']]
Y = movies_with_budget.DomesticTotalGross
linmodel = sm.OLS(Y, X).fit()
linmodel.summary()
```

```python
plt.scatter(movies_with_budget.Budget, movies_with_budget.DomesticTotalGross)
plt.xlabel('Budget')
plt.ylabel('Box Office')
plt.plot(movies_with_budget.Budget, linmodel.predict(), 'rs')
```

### Include Rating (Categorical Variables) in the model

```python
m = movies_with_budget
pd.get_dummies(m['Rating'])
```

```python
m = pd.concat([m,pd.get_dummies(m['Rating'])], axis=1)
m.head()
```

```python
X = m[['Budget', 'PG', 'PG-13', 'Ones']]
Y = m['DomesticTotalGross']
rating_linmodel = sm.OLS(Y, X).fit()
rating_linmodel.summary()
```

```python
# plot the predictions 
plt.plot(m['DomesticTotalGross'], rating_linmodel.predict(X), 'go')
plt.title('Over/Underprediction of the Budget + Rating Model')
plt.xlabel('Actual Gross')
plt.ylabel('Predicted Gross by Model')

# 1:1 line
plt.plot(m['DomesticTotalGross'], m['DomesticTotalGross'], 'm--', label = 'Ideal Prediction')
plt.legend()
```
